21. CLASSIFY MOVIES BASED ON RATINGS [EXCELLENT, GOOD, AND AVERAGE]

```python
newdata.columns
```

```python
newdata.head(5)
```

```python
# We are creating our own function to classify Movies based on Ratings.

# def rating - function name
# (rating) - let me pass argument

def rating (rating):
    if rating >= 7:
        return 'Excellent'
    elif rating >= 6:
        return "Good"
    else:
        return "Average"
```

```python
#lets create new column for this newly created category. "data['rating_cat']="

newdata['rating_cat'] = newdata['Rating'].apply(rating) 
```

23. FIND UNIQUE VALUES FROM GENRE

```python
newdata.Genre
```

```python
# we are going to perform three steps to find unique values in Genre column
# 1. Split items with commas
# 2. Convert this 2D list to 1D list to find unique values
# 3. Find unique value in this empty list
```

```python
# Step 1. Split items with commas

list1 = []
for value in newdata['Genre']:
    list1.append(value.split(','))

#Here we split items by comma. Then append this value over empty list
```

```python
list1
```

```python
# Step 2. In this second step, 2D list is converted to 1D list.
# # lets create this empty list for one_d

one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
# Step 3 : find unique value in this empty list

uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
